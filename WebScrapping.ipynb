```python
import pandas as pd
```

```python
week_num = "2"
week = "week_num_min="+week_num+"&week_num_max="+week_num
url2 ="https://www.pro-football-reference.com/play-index/pgl_finder.cgi?request=1&match=game&year_min=2019&year_max=2019&season_start=1&season_end=-1&pos%5B%5D=QB&pos%5B%5D=WR&pos%5B%5D=RB&pos%5B%5D=TE&pos%5B%5D=OL&pos%5B%5D=DL&pos%5B%5D=LB&pos%5B%5D=DB&is_starter=E&game_num_min=0&game_num_max=99&"+ week +"&c1stat=pass_att&c1comp=gt&c1val=1&c5val=1.0&order_by=pass_rating"
df = pd.read_html(url2)[0]
df.columns = df.columns.map('|'.join).str.strip('|')
df = df[df["Unnamed: 0_level_0|Rk"]!= "Rk"]
df["Passing|Cmp"]= df["Passing|Cmp"].astype(float)
df["Passing|Att"]= df["Passing|Att"].astype(float)
df["Passing|Yds"]= df["Passing|Yds"].astype(float)
df["Passing|TD"]= df["Passing|TD"].astype(float)
df["Passing|Int"]= df["Passing|Int"].astype(float)
df["Passing|Sk"]= df["Passing|Sk"].astype(float)
df = df.drop(["Unnamed: 1_level_0|Player", "Unnamed: 0_level_0|Rk","Unnamed: 2_level_0|Pos","Unnamed: 5_level_0|Lg","Unnamed: 4_level_0|Date","Unnamed: 3_level_0|Age"],axis=1)
df = df.groupby("Unnamed: 6_level_0|Tm").agg({"Unnamed: 7_level_0|Unnamed: 7_level_1":"first", "Unnamed: 8_level_0|Opp":"first","Unnamed: 9_level_0|Result":"first","Unnamed: 11_level_0|Week":"first","Unnamed: 12_level_0|Day":"first","Passing|Cmp":"sum","Passing|Att":"first","Passing|Sk":"sum","Passing|Yds":"sum"})

df
```

```python
def team_stat_by_week(df):
    df.columns = df.columns.map('|'.join).str.strip('|')
    df = df.drop(["Unnamed: 1_level_0|Player", "Unnamed: 0_level_0|Rk","Unnamed: 2_level_0|Pos","Unnamed: 5_level_0|Lg","Unnamed: 4_level_0|Date","Unnamed: 3_level_0|Age"],axis=1)
    df = df.groupby("Unnamed: 6_level_0|Tm").agg({"Unnamed: 7_level_0|Unnamed: 7_level_1":"first", "Unnamed: 8_level_0|Opp":"first","Unnamed: 9_level_0|Result":"first","Unnamed: 11_level_0|Week":"first","Unnamed: 12_level_0|Day":"first","Passing|Cmp":"sum","Passing|Att":"first","Passing|Sk":"sum","Passing|Yds.1":"sum"})
    df = df.reset_index()
    return df
```

```python
def week_iterating():
    for x in range(18)[1:18]:
        week = "week_num_min="+str(x)+"&week_num_max="+str(x)
        url2 ="https://www.pro-football-reference.com/play-index/pgl_finder.cgi?request=1&match=game&year_min=2019&year_max=2019&season_start=1&season_end=-1&pos%5B%5D=QB&pos%5B%5D=WR&pos%5B%5D=RB&pos%5B%5D=TE&pos%5B%5D=OL&pos%5B%5D=DL&pos%5B%5D=LB&pos%5B%5D=DB&is_starter=E&game_num_min=0&game_num_max=99&"+ week +"&c1stat=pass_att&c1comp=gt&c1val=1&c5val=1.0&order_by=pass_rating"
        df = pd.read_html(url2)[0]
        if x == 1:
            passing = team_stat_by_week(df)
        else:
            df = team_stat_by_week(df)
            passing = passing.append(df)
    return passing
            
passing = week_iterating()
passing
```
